-1 ~ +1 , 0.001

y = x^2 + sin(x)**굵은 텍스트**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import tensorflow as tf
from tensorflow import keras
```

```python
from tensorflow.keras import optimizers
from tensorflow.keras.layers import Dense, Input
```

```python
import time
```

```python
x=np.arange(-1,1,0.001)
np.random.shuffle(x)
y = x**2 + np.sin(x*10)

plt.scatter(x,y)
plt.show()
```

```python

split_index = int(x.shape[0]*0.8)
print(split_index)

train_x, test_x = x[:split_index], x[split_index:]
train_y, test_y = y[:split_index], y[split_index:]

plt.scatter(train_x, train_y)
plt.scatter(test_x, test_y, color='r')
```

# Optimizer

## Adam

```python
model = keras.Sequential()
model.add(Input(1))
model.add(Dense(10, activation='tanh'))
model.add(Dense(10, activation='tanh'))
model.add(Dense(1))


model.compile(optimizer="Adam", loss="mse")
# SGD, AdaGrad, RMSProp


hist = model.fit(train_x, train_y, epochs=1000, verbose=0, validation_split=0.1)

```

```python
# 훈련 손실값과 검증 손실값
plt.plot(hist.history['loss'])
plt.plot(hist.history['val_loss'])
plt.legend()
```

```python
plt.plot(hist.history['loss'],label='loss')
plt.plot(hist.history['val_loss'],label='val_loss')
plt.legend()
```

### batch_size

연산 한 번에 들어가는 데이터의 크기

배치사이즈가 너무 큰 경우 한 번에 처리해야할 데이터의 양이 많아지므로, 학습 속도가 느려지고, 메모리 부족 문제가 발생할 수 있음.

배치 사이즈가 너무 적은 경우 데이터를 대상으로 가중치를 업데이트 하고, 이 업데이트가 자주 발생하므로, 훈련이 불안정해질 수 있음.

```python
model = keras.Sequential()
model.add(Input(1))
model.add(Dense(10, activation='relu'))
model.add(Dense(10, activation='relu'))
model.add(Dense(1))


model.compile(optimizer="RMSProp", loss="mse")
# SGD, AdaGrad, RMSProp


hist = model.fit(train_x, train_y, epochs=10000, verbose=0,batch_size=128, validation_split=0.1)

plt.plot(hist.history['loss'],label='loss')
plt.plot(hist.history['val_loss'],label='val_loss')
plt.legend()
```

```python
y_ =model.predict(test_x)
plt.scatter(test_x,test_y)
plt.scatter(test_x, y_)
plt.show()
```

```python
model = keras.Sequential()
model.add(Input(1))
model.add(Dense(10, activation='relu'))
model.add(Dense(10, activation='relu'))
model.add(Dense(1))


model.compile(optimizer="RMSProp", loss="mse", metrics=["mae"])
# SGD, AdaGrad, RMSProp


hist = model.fit(train_x, train_y, epochs=1000, verbose=0,batch_size=128, validation_split=0.1)

plt.plot(hist.history['loss'],label='loss')
plt.plot(hist.history['val_loss'],label='val_loss')
plt.legend()

loss, mae = model.evaluate(test_x, test_y)
print("loss = ",loss)
print("mae = ", mae)
```

# 다시 한 번 해보기

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
import tensorflow as tf
from tensorflow.keras import optimizers

from  tensorflow import keras
from tensorflow.keras.layers import Dense, Input

import time
```

```python
x = np.arange(-1,1,0.001)
np.random.shuffle(x)
y = x**2 + np.sin(x*10)

plt.scatter(x,y)
plt.show()
```

```python
index = int(len(x)*0.8)

train_x, test_x = x[:index],x[index:]
train_y, test_y = y[:index],y[index:]

plt.scatter(train_x, train_y)
plt.show()
plt.scatter(test_x, test_y)
plt.show()
```

```python
model = keras.Sequential()
model.add(Input(1))
model.add(Dense(10, activation="tanh"))
model.add(Dense(10, activation="tanh"))
model.add(Dense(1))

model.compile(loss = "mse", optimizer="Adam", metrics=["mae"])

hist = model.fit(train_x, train_y, epochs=1000, batch_size=128, validation_split=0.1)

plt.plot(hist.history['loss'], label='loss')
plt.plot(hist.history['val_loss'], label='val_loss')
plt.legend()

loss, mae = model.evaluate(test_x, test_y)
print("loss = ",loss)
print("mae = ", mae)

y_ = model.predict(test_x)
plt.scatter(test_x, test_y)
plt.scatter(test_x, y_)
plt.show()
```
